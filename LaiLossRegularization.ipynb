```python
import pandas as pd

import torch
from torch.utils.data import DataLoader, TensorDataset

from fedot.core.data.data import InputData, OutputData
from fedot.core.repository.tasks import Task, TaskTypesEnum
from fedcore.models.base_models.mlp import MLP
from fedcore.models.network_modules.losses import LaiLoss
```

```python
torch.manual_seed(42)
```

```python
def convert_df_to_dl(X_train, y_train: pd.DataFrame,
                     batch_size: int = 128,
                     shuffle: bool = True):
    X_train = X_train.to_numpy()
    y_train = y_train.to_numpy()
    X_train = torch.tensor(X_train, dtype=torch.float32)
    y_train = torch.tensor(y_train, dtype=torch.float32).unsqueeze(-1)  # !!
    train_ds = TensorDataset(X_train, y_train)
    return DataLoader(train_ds, batch_size, shuffle=shuffle)
```

```python
# Get data
train = pd.read_csv('examples/data_examples/kaggle_s3e1/train.csv')
X_test = pd.read_csv('examples/data_examples/kaggle_s3e1/test.csv')

train.drop(columns='id', inplace=True)
X_test.drop(columns='id', inplace=True)

X_train, y_train = train.drop(columns='MedHouseVal'), train['MedHouseVal']

# Convert to data loader
train_dl = convert_df_to_dl(X_train, y_train)
X_test = torch.tensor(X_test.to_numpy(), dtype=torch.float32)
```

```python
# Base MLP model
mlp_model = MLP(input_size=X_train.shape[1], output_size=1)
mlp_model.fit(num_epochs=1000, train_dl=train_dl, plot_loss=True)
y_pred = mlp_model.predict(X_test)
```

```python
submission = pd.read_csv('examples/data_examples/kaggle_s3e1/sample_submission.csv')
submission['MedHouseVal'] = y_pred
submission.to_csv('submission.csv', index=False)
```
